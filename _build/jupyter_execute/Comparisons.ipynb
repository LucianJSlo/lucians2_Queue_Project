# Comparisons between Methods

We did a lot of comparisons when it came to this topic. Not only did we compare "Match" Queues and "Fill" Queues, but we also compared their performance on a uniform/poisson distribution, we calculated wait times by party/individual, we compared each queue's mean/median wait time, and we saw how many people in our queue were processed. With all this information, let's pit the two against each other and find out who comes out on top

### Individual vs. Party

- **Match Queues** generally have a **Lower Individual** Wait Time. This means if you asked every player how long they wait to start a game, they will respond with a lower time than if you saw how long a team got processed.

- **Fill Queues** had a **Higher Individual Uniform** Wait Time, where as the wait time for **Poisson was Equal**. This implies that if you ask each player how long they wait to start a game, more people will say they waited longer than if you saw which teams got processed. This implies that large teams are taking longer to get processed.

### Mean vs. Median

**All Trials had a Smaller Median than Mean** !!

The medians (from smallest to largest) are as follows:

- **Match Uniform**, **Match Poisson**, **Fill Poisson**, **Fill Uniform**

The means (from smallest to largest) are as follows:

- **Fill Poisson**, **Match Poisson**, **Match Uniform**, **Fill Uniform**

It's important to note that although Match Unifrom had the lowest median wait time (~2), its mean wait time was nearly 10 times that (~20). Fill Poisson on the other hand, has a nearly equal median and mean. 

This shows that when players join the queue with a party size of a poisson distribution, the Fill Queue becomes the most consistent way of grouping players. 

Match queues, on the other hand, are the fastest when under a uniform party spread, but about 25% of players will have to wait more than 5 times that, up to almost 250 times 

### Uniform vs. Poisson

- Under **Uniform Conditions**, A **Match Queue**, for most people, will generally be **much faster** than a **Fill Queue**; However, **Match Queues** are also **Right Skewed** so they will generally have a **Higher Standard Deviation** than **Fill Queues**

- Under **Poisson Conditions** A **Fill Queue** consistently **perfroms better** than a **Match Queue**. Even though a **Match Queue** may have the faster average median, it leaves **a lot of players unpaired!!**

### Players Unpaired

The average players left over (from smallest to largest) is as follows:
- **Fill Poisson**, **Match Uniform**, **Fill Uniform**, **Match Poisson**

## The Verdict:

Our two standout performances were Match Queues on Unifrom with it's lighting quick matching and Fill Queues on Poisson with stable consistency. All that's left is to determine which distribution matches the real world: do most people play by themselves, with a few dedicated duos/trios? Or is the player party size pool more equally spread out? Spoiler alert, most people play by themselves, especially since it could be more difficult getting a friend or 2 or 3 to play with you at the same time as well. 

Knowing this, the **Most Efficient Queue** to implement in terms of **player pairing and pairing speed** would be... **The Fill Queue!!**
